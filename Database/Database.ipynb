#### N:M Datenbank

In diesem Ansatz werden die Autoren in einer separaten Tabelle gespeichert. Mittels einer dritten Tabelle wird dann die Beziehung zwischen der Autoren und der Arbeiten gespeichert, sodass sich eine N:M Beziehung entwickelt.

##### author Tabelle

Hier wird die Autoren Tabelle erstellt. Hierbei gibt es zu beachten das die `author_name` `UNIQUE` sein muss. Dies verhindert, dass später der gleiche Autor mehrfach in der Datenbank steht. Dadurch kann dem Autor dann eine einheitliche `author_ID` zugewiesen werden.

```python
def create_author(database):
    with sqlite3.connect(database) as con:
        cur = con.cursor()
        
        sql="CREATE TABLE author ("
        sql +="author_ID INTEGER PRIMARY KEY AUTOINCREMENT UNIQUE,"
        sql +="author_name TEXT UNIQUE)"
        
        cur.execute(sql)
```

Hier wird die Tabelle für die Titel und die Abstract erstellt. Auch hier ist das einzige wichtige, dass sowohl `title` als auch `abstract` nicht leer sind.

```python
def create_thesis_english(database):
    with sqlite3.connect(database) as con:
        cur = con.cursor()
        
        sql="CREATE TABLE thesis_english ("
        sql +="thesis_en_ID INTEGER PRIMARY KEY AUTOINCREMENT UNIQUE,"
        sql +="title TEXT NOT NULL,"
        sql +="abstract TEXT NOT NULL)"
        
        cur.execute(sql)
```

```python
create_thesis_english(database)
```

##### relation Tabelle

Die letzte Tabelle ist da um die Beziehung zu beschreiben. Hierbei werden die IDs der `author`-Tabelle mit dem IDs der `thesis_english`-Tabelle verbunden. Das wird mittels eines `FOREIGN KEY` gemacht.

```python
def create_relation(database):
    with sqlite3.connect(database) as con:
        cur = con.cursor()
        
        sql="CREATE TABLE relation ("
        sql +="thesis_en_ID INTEGER,"
        sql +="author_ID INTEGER,"
        sql +="FOREIGN KEY('author_ID') REFERENCES 'author'('author_ID'),"
        sql +="FOREIGN KEY('thesis_en_ID') REFERENCES 'thesis_english'('thesis_en_ID'))"
        
        cur.execute(sql)
```

```python
create_relation(database)
```

### Hilfsfunktionen für das Einlesen der Daten aus der JSON Datei 

#### Einfaches Einfügen 
In diesem Abschnitt geht es um das Einfache Einfügen der JSON-Daten in die Datenbank. Damit ist gemeint, dass der Aufbau der JSON Datei einfach übernommen wird. Heißt jedes Element in der JSON Datei wird ein Feld einer Tabelle ergeben.


Die Funktion sql_insert_data() generiert die INSERT INTO Statements.
Hierbei werden die Dictianary-Objekte (zb. `data_items_de` und `data_items_en`) in einer Tabelle eingefügt.
Dabei soll aus einem Dict der Form

     {"title": string, "abstract": string, "author": string}
     
automatisch ein INSERT-Statement generiert und ausgeführt werden.
Das INSERT-Statement soll dann wie folgend aussehen:

    INSERT INTO table (title, abstract, author) VALUES (?,?,?) 
    
Die Fragezeichen sind Platzhalter (Bind-Parameters), die direkt dem Cursor für das Execute mit übergeben werden können. So kann Typsicherheit beim Einfügen gewährleistet werden und man muss sich nicht mit dem Escapen von Anführungszeichen und Newlines bei Strings beschäftigen.
    
Die Felder (keys) im Dictionary müssen natürlich alle als Spalten in der Datenbank vorhanden sein. Geschachtelte Werte sind nicht zulässig.


```python
def sql_insert_data(database, table, data_items):
 
    with sqlite3.connect(database) as con:
        cur = con.cursor()

        cnt = 0
        for item in data_items:

            # generate list of column names and placeholders
            fields = item.keys()
            record = list(item.values())
            params = len(fields)*['?']

            fields = ",".join(fields)
            params = ",".join(params)

            sql = "INSERT INTO %s (%s) VALUES (%s)" % (table, fields, params)
            
            try:
                cur.execute(sql, record)
                cnt += 1
            except Exception as e:
                print("EXCEPTION: %s " % e)
                print(sql)
                print(item)
                raise

        print("%d records inserted into %s." % (cnt, table))
```

In dieser Funktion wird die Logik zum Laden der JSON-Sätze um zum Einfügen in die SQLite-Datenbank gemacht.

```python
def sql_load_json(json_file, database, table):
    # read data items from file
    data_items = []
    with open(json_file, encoding="utf-8") as file:
        for line in file:
            item = json.loads(line)
            data_items.append(item)

    # insert data_items into database table
    sql_insert_data(database, table, data_items)
```

### Komplexeres Datenbankschema für die Englische Datenbank

Hier werden die Titel und Abstracts in einer Tabelle gespeichert und die Autoren in einer anderen Tabellen. Die beiden Tabellen werden dann mittels Fremdschlüssel zu einer n:m Beziehung gemacht. In der Absicht das sich dadurch das suchen und sortieren nach Autoren vereinfachen lässt. Da nun nicht mehr die Gesamte Zeile der JSON-Datei verwendet werden kann, muss sich auch das `INSERT`-Statement etwas verändern.

#### Einfügen der Titel und Autoren

Die Funktion `sql_insert_thesis_enlisch(database, data_items_en)` speichert Titel und Abstract in eine Tabelle ein. Dabei werden die Autoren in der JSON-Datei ignoriert. Die Autoren werden in einer separaten Funktion eingefügt.

```python
def sql_insert_thesis_english(database, data_items_en):
    with sqlite3.connect(database) as con:
        cur = con.cursor()
        
        cnt = 0
        for item in data_items_en:
            sql = ("INSERT INTO thesis_english (title,abstract)"
                   "VALUES (?, ?)"
            )
            data = (item['title'], item['abstract'])
            
            try:
                cur.execute(sql, data)
                cnt += 1
            except Exception as e:
                print("EXCEPTION: %s " % e)
                print(sql)
                #print(item)
                print(data[1])
                raise
        print("%d records inserted into thesis_english." % (cnt))
        
```

Der nächste Codebereich ruft die `sql_insert_thesis_enlisch(database, data_items_en)` auf.

```python
sql_insert_thesis_english(database, data_items_en_2)
```

Die Funktion `sql_insert_author(database, data_items_en)`holt sich die Autoren aus der JSON Datei und speichert sie in die Autoren Tabelle ein. Bei diesen `INSERT` ist es wichtig das darauf geachtet wird das ein Autor nicht doppelt vorkomme. Das wird sichergestellt mit dem `INSERT OR IGNOR`. Bei `IGNOR` wir dein Wert der schon in der Datenbank steht einfach übersprungen. Da beim Erstellen der Tabelle für die Autoren auch noch darauf geachtet wurde das das Element `author` `UNIQUE` ist, würde zusätzlich eine Fehlermeldung kommen, sollte ein Autor doppelt gespeichert werden.

```python
def sql_insert_author(database, data_items_en):
    with sqlite3.connect(database) as con:
        cur = con.cursor()
        
        cnt = 0
        for item in data_items_en:
            for author in item['author']:
                sql = ("INSERT OR ignore INTO author (author_name)"
                       "VALUES (?)")
                data = (author)

                try:
                    cur.execute(sql, (data,))
                    cnt += 1
                except Exception as e:
                    print("EXCEPTION: %s " % e)
                    print(sql)
                    #print(item)
                    print(data[1])
                    raise
        print("%d records inserted into author." % (cnt))
        
```

Der nächste Codebereich ruft die `sql_insert_author(database, data_items_en)` auf.

```python
sql_insert_author(database, data_items_en_2)
```

Die Nächste Funktion verknüpft die `thesis_english`-Tabelle mit der `author`-Tabelle. Hierbei wird wieder die JSON Datei geladen. Mittels einer FOR-Schleife wird dann Zeile für Zeile durch die JSON-Datei durchgegangen. Dabei wird sich dann der Titel der jeweiligen Zeile gemerkt. Dieser Titel wird dann dafür verwendet um das Richtige Element aus der `thesisi_englisch`-Tabelle zu holen. Die zweite FOR-Schleife wird dazu verwendet die Autoren durch zu gehen. Mittels dieser Autoren wird geschaut welche `author_ID` dann zu welcher `thesis_ID` gehören. Die Beziehung der Beiden Tabellen wird dann in der `relation`-Tabellen gespeichert.

```python
def sql_insert_relation(database, data_items_en):
    with sqlite3.connect(database) as con:
        cur = con.cursor()
        
        cnt = 0
        for item in data_items_en:
            for author in item['author']:
                sql = ("INSERT INTO relation (thesis_en_ID, author_ID)"
                       "SELECT th.thesis_en_ID, ath.author_ID "
                       "FROM "
                       "(SELECT thesis_en_ID FROM thesis_english WHERE title LIKE (?)) AS th "
                       "CROSS JOIN "
                       "(SELECT author_ID FROM author WHERE author_name LIKE (?)) AS ath ")
                data = (item['title'], author)

                try:
                    cur.execute(sql, (data))
                    cnt += 1
                except Exception as e:
                    print("EXCEPTION: %s " % e)
                    print(sql)
                    #print(item)
                    print(data[1])
                    raise
        print("%d records inserted into author." % (cnt))
```

```python
sql_insert_relation(database, data_items_en_2)
```

##### n:m Tabellen Aufbau

Bei diesen Aufbau ist zu beachten das zuerst die jeweiligen Verbindungen, des zu löschenden Elements, aus der `relation`-Tabelle gelöscht werden muss, bevor die Werte aus der `thesis_english`-Tabelle gelöscht werden kann.

###### löschen in der relation Tabelle

Der erste Schritt hier ist es die Werte aus der relation Tabelle zu löschen.


```python
def sql_delete_relation(database, field, value):

    with sqlite3.connect(database) as con:
        cur = con.cursor()
      
        sql = ("DELETE FROM relation WHERE thesis_en_ID IN (SELECT thesis_en_ID FROM thesis_english WHERE %s = (?))") % (field)
        data = (value)

        cur.execute(sql, (data,))
        num_rows = cur.rowcount
        
        print("%d records deleted from relation." % (num_rows))
```
