### deutscher Datensatz

Synchron zum englischen Datensatz wird hier nun den Deutsche Datensatz geholt. 

```python
query_all_de = """SELECT * FROM thesis_german"""

df_all_de = pd.read_sql_query(query_all_de, conn)
df_all_de
```

```python
def save_to_json_abstract(df_query_lenght, df_query, save_path, number):
    count = 0    
    train = []
    vali = []
    test = []

    number2 = df_query_lenght
    seventyPercent = round(number2 * 0.7)
    twentyPercent = round(number2 * 0.2)
    tenPercent = round(number2 * 0.1)
                  
    for sentence in df_query.abstract:
        
        temp_data = {}
        temp_data['src'] = []
        temp_data['tgt'] = []

        nltk_sent_token = nltk.sent_tokenize(sentence)      

        for word in nltk_sent_token:
            nlkt_word = nltk.word_tokenize(word)

            temp_data['src'] += [nlkt_word]         
                       
        nlkt_sent_token2 = nltk.sent_tokenize(df_query.title[count])
        
        for sentence2 in nlkt_sent_token2:
            nlkt_word2 = nltk.word_tokenize(sentence2)

            temp_data['tgt'] += [nlkt_word2]               
        
        if count < seventyPercent:
            train.append(temp_data)
            
        if count >= seventyPercent and count < (seventyPercent + twentyPercent):
            vali.append(temp_data)  
            
        if count >= (seventyPercent + twentyPercent) and count > tenPercent:
            test.append(temp_data)
      
        json_train = json.dumps(train, indent=2)
        json_vali = json.dumps(vali, indent=2)
        json_test = json.dumps(test, indent=2)
        count +=1
     
    with open(save_path + ".train.0.json", "w") as f, open(save_path + ".valid.0.json", "w") as g, open(save_path + ".test.0.json", "w") as h:
        f.write(json_train)
        g.write(json_vali)
        h.write(json_test)
```

```python
save_to_json_abstract(4199, df_all_eng, 'BertSum/json_data/english/english', '0')
```

```python
save_to_json_abstract(20, df_test, 'BertSum/json_data/test_data2/test', '0')
```

```python
save_to_json_abstract(939, df_all_de, 'BertSum/json_data/german/german', '0')
```
