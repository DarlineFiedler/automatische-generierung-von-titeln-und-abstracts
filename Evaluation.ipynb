### WER Durchführung
In diesem Abschnitt wird WER Durchgeführt. Hierzu wird das Python-Pakt JiWER genutzt welches die einfache Berechnung von WER ermöglicht.

#### Daten Einlesen
Es werden zwei Dateien eingelesen. Bei der [...]_gold.json handelt es sich um mehrere original Titel der Interspeech Abstractbooks. In der dazugehörigen [...]_pred.json stehen die von BERTSUM generierten Titel.


```
english_gold = 'automatische-generierung-von-titeln-und-abstracts/BertSum/results/english-test/english_gold.json'

data_items_gold = []

with open(english_gold) as file_gold:
    for line_gold in file_gold:
      data_items_gold.append(json.loads(line_gold))
```

```
english_pred = 'automatische-generierung-von-titeln-und-abstracts/BertSum/results/english-test/english_pred.json'

data_items_pred = []

with open(english_pred) as file_pred:
    for line_pred in file_pred:
      data_items_pred.append(json.loads(line_pred))
```

```

all_gold = ""
first_gold= ""
for item in data_items_gold:
  counter = 0
  for titel in item['Gold']:
    if (counter == 0):
      first_gold = titel
    all_gold=all_gold + ' ' + titel
    counter += 1

all_pred = ""
first_pred = ""
for item in data_items_pred:
  counter = 0
  for titel in item['Pred']:
    if (counter == 0):
      first_pred = titel
    all_pred=all_pred + ' ' + titel
    counter += 1

wer_all_english = wer(all_gold, all_pred)
print(wer_all_english)

wer_one_english = wer(first_gold, first_pred)
print(wer_one_english)

```

```
ground_truth = ["Automatic Evaluation of Children Reading Aloud on Sentences and Pseudowords"]
hypothesis = ["To predict these scores automatically, features based on reading speed and number of disfluencies were extracted, after an automatic disfluency detection. Various regression models were trained, with Gaussian process regression giving best results for automatic features. A computational tool that assists with recording reading tasks, automatically analyzing them and providing performance metrics could be a significant help."]

error = wer(ground_truth, hypothesis)
print(error)
```
